# ðŸ“Š Sales Data Analysis for a Retail Company
This project analyzes one year of fictional sales data for a retail company. It includes trends, visualizations, and insights for business decision-making.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
df.info()
```

```python
# Total sales per month
monthly_sales = df.groupby('Month')['Sales'].sum().reindex(
    ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)
monthly_sales.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('Total Sales by Month')
plt.ylabel('Revenue ($)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total sales by city
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
city_sales.plot(kind='bar', figsize=(10,6), color='orange')
plt.title('Sales by City')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top-selling products
product_sales = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
product_sales.plot(kind='bar', figsize=(10,6), color='green')
plt.title('Sales by Product')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.show()
```

### ðŸ“Œ Key Insights
- **Highest revenue month**: Easily visible from the bar chart
- **Top city**: Best-performing location for the year
- **Best-selling product**: Focus for future marketing or bundling
- This analysis provides a foundation for data-driven decision making.
